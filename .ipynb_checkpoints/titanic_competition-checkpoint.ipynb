<h1>Check and Merge Data Set</h1><br>
data Pre-Processing

```python
import pandas as pd
import numpy as np

test_set = pd.read_csv('test.csv')
train_set = pd.read_csv('train.csv')

gender_submission = pd.read_csv('gender_submission.csv')
test_set = pd.merge(test_set, gender_submission, how='inner', on='PassengerId')

# Resorting Columns
c_num1 = test_set.columns[0:1].to_list()   # PassengerId
c_num2 = test_set.columns[-1:].to_list()     # Survived
c_others = test_set.columns[1:-1].to_list()   # rest columns

# resort
test_set = test_set[c_num1+c_num2+c_others]

print('test_set')
display(test_set)

print('\n\ntrain_set')
display(train_set)
```

```python
# Check Correlation

corr = train_set.corr().abs().nlargest(3, 'Survived').drop('Survived')['Survived']
display(corr)
# Result : Pclass(0.338481), Fare(0.257307)

# Set train set
x = pd.DataFrame(np.c_[train_set[corr.index[0]], train_set[corr.index[1]]], columns = ['Pclass','Fare'])
y = train_set['Survived']

print(x)

# Visualize with Scatter Plot

import matplotlib.pyplot as plt

# Pclass vs Survived
plt.subplot(121)  # 1í–‰ 2ì—´ 1ë²ˆì°¨íŠ¸
plt.scatter(x.iloc[:,0], y)
plt.xlabel("Pclass")
plt.ylabel("Survived")
plt.title("Pclass vs Survived")

# Fare vs Survived
plt.subplot(122)  # 1í–‰ 2ì—´ 2ë²ˆì°¨íŠ¸
plt.scatter(x.iloc[:,1], y)
plt.xlabel("Fare")
plt.ylabel("Survived")
plt.title("Fare vs Survived")
plt.show()
```

<h2>Logistic Regression</h2>

```python
from sklearn import linear_model

# Logistic Regression
log_regress = linear_model.LogisticRegression()
log_regress.fit(X = x, y = y)

print('\n--Logistic Modeling--')
print('Logistic Intercept:',log_regress.intercept_[0])
print('Logistic Coefficient:',log_regress.coef_[0][0])


# drop NaN included data(1)
test = pd.DataFrame(np.c_[test_set['Pclass'], test_set['Fare'],test_set['Survived']],columns = ['Pclass','Fare','Survived']).dropna()

# Predict with test data set
preds = log_regress.predict(test[['Pclass','Fare']])

#---generate table of predictions vs actual---
print("---Confusion Matrix---")
print(pd.crosstab(res, test['Survived']))

#---get the accuracy of the prediction---
print("---Accuracy---")
print(log_regress.score(X = test[['Pclass','Fare']], y = test['Survived']))

from sklearn import metrics

# View summary of common classification metrics
print("---Metrices---")
print(metrics.classification_report(y_true = test['Survived'], y_pred = preds))
```
